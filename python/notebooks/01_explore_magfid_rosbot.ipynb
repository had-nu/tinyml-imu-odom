```python
"""
TL;DR: para esta fase manteremos apenas timestamp(Time_Sample), aceleração (Acc_X, Acc_Y, Acc_Z) e giroscópio (Gyro_X, Gyro_Y, Gyro_Z).
========================
Os dados MRU possuem várias colunas, mas apenas algumas são relevantes para o modelo TinyML. 
A coluna Unnamed: 0 é apenas um índice salvo no CSV e deve ser removida. 
Time_Sample [sec] é o timestamp em segundos e serve para medir a taxa de amostragem e organizar janelas temporais. 
As colunas de GNSS (Lat_GNSS, Long_GNSS, Height_GNSS) representam latitude, longitude e altitude, mas não serão usadas porque GNSS não é útil para odometria indoor neste projeto. 
As colunas Acc_X, Acc_Y e Acc_Z trazem a aceleração nos três eixos em unidades de “g” e são fundamentais para identificar movimento linear, inclinação e impactos. 
Já Gyro_X, Gyro_Y e Gyro_Z medem a rotação em graus por segundo; especialmente Gyro_Z, que indica curvas no plano horizontal e é essencial para estimar Δθ. 
As colunas do magnetômetro (Magn_X, Magn_Y, Magn_Z) capturam o campo magnético, mas não serão usadas na Fase 1 por causa do alto ruído e baixa utilidade para janelas curtas. 


| Coluna            | Unidade | Tipo         | Uso no Projeto | Descrição                   |
| ----------------- | ------- | ------------ | -------------- | --------------------------- |
| Unnamed: 0        | —       | índice lixo  | remover        | índice automático sem valor |
| Time_Sample [sec] | sec     | timestamp    | usar           | tempo de amostragem         |
| Lat_GNSS [deg]    | grau    | posição GNSS | descartar      | latitude do robô            |
| Long_GNSS [deg]   | grau    | posição GNSS | descartar      | longitude do robô           |
| Height_GNSS [m]   | metro   | posição GNSS | descartar      | altitude                    |
| Acc_X [g]         | g       | aceleração   | usar           | aceleração eixo X           |
| Acc_Y [g]         | g       | aceleração   | usar           | aceleração eixo Y           |
| Acc_Z [g]         | g       | aceleração   | usar           | aceleração eixo Z           |
| Gyro_X [dps]      | °/s     | giroscópio   | usar           | rotação eixo X              |
| Gyro_Y [dps]      | °/s     | giroscópio   | usar           | rotação eixo Y              |
| Gyro_Z [dps]      | °/s     | giroscópio   | usar           | rotação eixo Z              |
| Magn_X [nT]       | nT      | magnetômetro | descartar¹     | campo magnético             |
| Magn_Y [nT]       | nT      | magnetômetro | descartar¹     | campo magnético             |
| Magn_Z [nT]       | nT      | magnetômetro | descartar¹     | campo magnético             |

\n¹ Magnetômetro pode ser usado em fases futuras para heading absoluto.
"""
```

```python
"""
Este bloco de teste de limpeza fará: 
Limpeza da estrutura do DataFrame, removendo colunas desnecessárias, deixando apenas timestamp + aceleração + giroscópio, e criando uma versão limpa e pronta para análise profunda.
"""

cols_drop = [
    "Unnamed: 0",
    "Lat_GNSS [deg]",
    "Long_GNSS [deg]",
    "Height_GNSS [m]",
    "Magn_X [nT]",
    "Magn_Y [nT]",
    "Magn_Z [nT]"
]

def load_n_clean(conf: str) -> pd.DataFrame:
    """
    Carrega TODOS os CSVs de uma Conf_X, concatena e remove colunas que
    não serão usadas (índice, GNSS, magnetômetro).
    """
    df = load_all_mru(conf)
    df_cleaned = df.drop(columns=cols_drop)
    df_cleaned.name = f"{conf}_clean"
    print(f"✓ {df_cleaned.name}: {df_cleaned.shape[0]:,} linhas x {df_cleaned.shape[1]} colunas (limpo)")
    return df_cleaned

df1_cleaned.head()
```
